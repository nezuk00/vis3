```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder, OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
import warnings
warnings.filterwarnings('ignore')

# Загрузка данных
try:
    df = pd.read_csv('Titanic-Dataset.csv')
except:
    # Если файл не найден, создадим синтетические данные для демонстрации
    print("Файл не найден. Создаю синтетические данные для демонстрации...")
    np.random.seed(42)
    n_samples = 891

    synthetic_data = {
        'PassengerId': range(1, n_samples + 1),
        'Survived': np.random.randint(0, 2, n_samples),
        'Pclass': np.random.randint(1, 4, n_samples),
        'Name': [f'Passenger {i}' for i in range(1, n_samples + 1)],
        'Sex': np.random.choice(['male', 'female'], n_samples),
        'Age': np.random.normal(30, 15, n_samples).clip(0, 80),
        'SibSp': np.random.randint(0, 5, n_samples),
        'Parch': np.random.randint(0, 4, n_samples),
        'Ticket': [f'Ticket_{i}' for i in range(1, n_samples + 1)],
        'Fare': np.random.exponential(50, n_samples).clip(0, 512),
        'Cabin': np.random.choice([f'C{i}' for i in 'ABCDEFG'] + [None], n_samples, p=[0.1]*7 + [0.3]),
        'Embarked': np.random.choice(['S', 'C', 'Q', None], n_samples, p=[0.7, 0.2, 0.08, 0.02])
    }

    df = pd.DataFrame(synthetic_data)

print("Первые 5 строк данных:")
print(df.head())
print("\nИнформация о данных:")
print(df.info())
print("\nСтатистика данных:")
print(df.describe(include='all'))
```

```python
# ВИЗУАЛИЗАЦИИ
def create_visualizations(df):
    """Создание визуализаций для анализа данных"""

    fig, axes = plt.subplots(2, 3, figsize=(18, 12))

    # 1. Распределение выживших
    sns.countplot(data=df, x='Survived', ax=axes[0,0])
    axes[0,0].set_title('Распределение выживших')

    # 2. Распределение по классам
    sns.countplot(data=df, x='Pclass', ax=axes[0,1])
    axes[0,1].set_title('Распределение по классам')

    # 3. Распределение по полу
    sns.countplot(data=df, x='Sex', ax=axes[0,2])
    axes[0,2].set_title('Распределение по полу')

    # 4. Распределение возраста
    sns.histplot(data=df, x='Age', bins=30, ax=axes[1,0])
    axes[1,0].set_title('Распределение возраста')

    # 5. Распределение стоимости билета
    sns.histplot(data=df, x='Fare', bins=30, ax=axes[1,1])
    axes[1,1].set_title('Распределение стоимости билета')

    # 6. Порт посадки
    sns.countplot(data=df, x='Embarked', ax=axes[1,2])
    axes[1,2].set_title('Порт посадки')

    plt.tight_layout()
    plt.show()

    # Корреляционная матрица для числовых признаков
    plt.figure(figsize=(10, 8))
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    correlation_matrix = df[numeric_cols].corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
    plt.title('Корреляционная матрица')
    plt.show()

create_visualizations(df)
```

```python
# АНАЛИЗ ПРОПУЩЕННЫХ ЗНАЧЕНИЙ
def analyze_missing_values(df):
    """Анализ пропущенных значений"""
    print("Анализ пропущенных значений:")
    missing_data = df.isnull().sum()
    missing_percent = (missing_data / len(df)) * 100

    missing_df = pd.DataFrame({
        'Количество пропусков': missing_data,
        'Процент пропусков': missing_percent
    })

    print(missing_df[missing_df['Количество пропусков'] > 0])

    # Визуализация пропущенных значений
    plt.figure(figsize=(10, 6))
    missing_data = missing_data[missing_data > 0]
    if len(missing_data) > 0:
        sns.barplot(x=missing_data.index, y=missing_data.values)
        plt.title('Пропущенные значения по признакам')
        plt.xticks(rotation=45)
        plt.show()
    else:
        print("Пропущенных значений не обнаружено!")

    return missing_df

missing_analysis = analyze_missing_values(df)
```

```python
# ОБНОВЛЕННАЯ ГЕНЕРАЦИЯ ИСКУССТВЕННЫХ ОШИБОК (единая версия)
def generate_artificial_errors(df):
    """Генерация сообщений об ошибках для случаев, когда предобработка не требуется"""
    error_messages = []

    # Проверка на отсутствие пропусков
    if df.isnull().sum().sum() == 0:
        error_messages.append("ОШИБКА: В данных отсутствуют пропущенные значения. Пропуск этапа заполнения пропусков.")

    # Проверка на полностью числовые данные
    if df.select_dtypes(include=['object']).empty:
        error_messages.append("ОШИБКА: Все данные уже являются числовыми. Пропуск этапа преобразования типов.")

    # Проверка на нормализованные данные
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    if len(numeric_cols) > 0:
        for col in numeric_cols:
            if abs(df[col].mean()) < 0.1 and abs(df[col].std() - 1) < 0.1:
                error_messages.append(f"ОШИБКА: Признак {col} уже нормализован. Пропуск этапа нормализации.")

    return error_messages

# Вывод искусственных ошибок (обновленный вызов)
artificial_errors = generate_artificial_errors(df)
if artificial_errors:
    print("Сгенерированные ошибки:")
    for error in artificial_errors:
        print(f"⚠️ {error}")
else:
    print("Искусственные ошибки не сгенерированы - все этапы предобработки требуются.")
```

```python
# ИСПРАВЛЕННЫЙ ПАЙПЛАЙН ПРЕДОБРАБОТКИ ДАННЫХ
from sklearn.preprocessing import OneHotEncoder

class DataPreprocessingPipeline:
    def __init__(self):
        self.preprocessor = None
        self.feature_names = None
        self.label_encoders = {}

    def create_pipeline(self, df, target_col='Survived'):
        """Создание пайплайна предобработки"""

        # Удаляем ненужные колонки
        columns_to_drop = ['PassengerId', 'Name', 'Ticket']
        df_processed = df.drop(columns=[col for col in columns_to_drop if col in df.columns], errors='ignore')

        # Разделяем признаки и целевую переменную
        X = df_processed.drop(columns=[target_col], errors='ignore')
        y = df[target_col] if target_col in df.columns else None

        # Определяем числовые и категориальные признаки
        numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
        categorical_features = X.select_dtypes(include=['object']).columns.tolist()

        print(f"Числовые признаки: {numeric_features}")
        print(f"Категориальные признаки: {categorical_features}")

        # Создаем трансформеры для числовых и категориальных признаков
        numeric_transformer = Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', StandardScaler())
        ])

        # Используем OneHotEncoder вместо LabelEncoder для категориальных признаков
        categorical_transformer = Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
        ])

        # Создаем ColumnTransformer
        self.preprocessor = ColumnTransformer(
            transformers=[
                ('num', numeric_transformer, numeric_features),
                ('cat', categorical_transformer, categorical_features)
            ])

        # Сохраняем имена признаков
        self.feature_names = numeric_features + categorical_features

        return self.preprocessor, X, y

    def fit_transform(self, X):
        """Обучение и преобразование данных"""
        if self.preprocessor is None:
            raise ValueError("Пайплайн не создан. Сначала вызовите create_pipeline()")

        X_transformed = self.preprocessor.fit_transform(X)

        # Получаем имена фичей после OneHot кодирования
        self.transformed_feature_names = self._get_feature_names()

        return X_transformed

    def transform(self, X):
        """Преобразование новых данных"""
        if self.preprocessor is None:
            raise ValueError("Пайплайн не создан. Сначала вызовите create_pipeline()")

        X_transformed = self.preprocessor.transform(X)
        return X_transformed

    def _get_feature_names(self):
        """Получение имен признаков после преобразования"""
        if self.preprocessor is None:
            return []

        feature_names = []

        # Для числовых признаков
        numeric_features = self.preprocessor.named_transformers_['num'].steps[0][1].feature_names_in_
        if numeric_features is not None:
            feature_names.extend(numeric_features)

        # Для категориальных признаков
        cat_transformer = self.preprocessor.named_transformers_['cat']
        if hasattr(cat_transformer, 'named_steps') and 'onehot' in cat_transformer.named_steps:
            onehot = cat_transformer.named_steps['onehot']
            categorical_features = cat_transformer.steps[0][1].feature_names_in_
            if hasattr(onehot, 'get_feature_names_out') and categorical_features is not None:
                cat_features = onehot.get_feature_names_out(categorical_features)
                feature_names.extend(cat_features)

        return feature_names

# Создание и применение пайплайна
pipeline = DataPreprocessingPipeline()
preprocessor, X, y = pipeline.create_pipeline(df)

print("\nДанные до предобработки:")
print(f"Размерность: {X.shape}")
print(f"Признаки: {list(X.columns)}")
print(f"Пропуски: {X.isnull().sum().sum()}")

# Применяем пайплайн
X_transformed = pipeline.fit_transform(X)

print(f"\nДанные после предобработки:")
print(f"Размерность: {X_transformed.shape}")
print(f"Тип данных: {type(X_transformed)}")
```

```python
# ДОПОЛНИТЕЛЬНАЯ ВИЗУАЛИЗАЦИЯ РЕЗУЛЬТАТОВ ПРЕДОБРАБОТКИ
def visualize_preprocessing_results(original_df, transformed_data, pipeline):
    """Визуализация результатов предобработки"""

    fig, axes = plt.subplots(2, 2, figsize=(15, 12))

    # 1. Сравнение распределений до и после стандартизации
    numeric_cols = original_df.select_dtypes(include=[np.number]).columns

    if len(numeric_cols) > 0:
        # Берем первый числовой признак для примера
        example_col = numeric_cols[0]
        axes[0,0].hist(original_df[example_col].dropna(), bins=30, alpha=0.7, label='До обработки', color='blue')

        # Находим соответствующий преобразованный признак
        if hasattr(pipeline, 'transformed_feature_names') and example_col in pipeline.transformed_feature_names:
            idx = pipeline.transformed_feature_names.index(example_col)
            if idx < transformed_data.shape[1]:
                axes[0,0].hist(transformed_data[:, idx], bins=30, alpha=0.7, label='После обработки', color='red')

        axes[0,0].set_title(f'Распределение {example_col} до и после обработки')
        axes[0,0].legend()

    # 2. Пропущенные значения до и после обработки
    missing_before = original_df.isnull().sum().sum()
    missing_after = 0  # После обработки пропусков не должно быть

    axes[0,1].bar(['До обработки', 'После обработки'], [missing_before, missing_after], color=['blue', 'green'])
    axes[0,1].set_title('Количество пропущенных значений')
    axes[0,1].set_ylabel('Количество пропусков')

    # 3. Статистики числовых признаков до обработки
    if len(numeric_cols) > 0:
        stats_before = original_df[numeric_cols].describe().loc[['mean', 'std']]
        im1 = axes[1,0].imshow(stats_before.values, cmap='viridis', aspect='auto')
        axes[1,0].set_title('Статистики до обработки (mean, std)')
        axes[1,0].set_xticks(range(len(numeric_cols)))
        axes[1,0].set_xticklabels(numeric_cols, rotation=45)
        axes[1,0].set_yticks([0, 1])
        axes[1,0].set_yticklabels(['mean', 'std'])
        plt.colorbar(im1, ax=axes[1,0])

    # 4. Статистики после обработки (только числовые признаки)
    if transformed_data.shape[1] > 0:
        # Берем только первые n числовых признаков для визуализации
        n_features = min(5, transformed_data.shape[1])
        stats_after = pd.DataFrame(transformed_data[:, :n_features]).describe().loc[['mean', 'std']]
        im2 = axes[1,1].imshow(stats_after.values, cmap='viridis', aspect='auto')
        axes[1,1].set_title('Статистики после обработки (первые 5 признаков)')
        axes[1,1].set_yticks([0, 1])
        axes[1,1].set_yticklabels(['mean', 'std'])
        plt.colorbar(im2, ax=axes[1,1])

    plt.tight_layout()
    plt.show()

# Визуализируем результаты
visualize_preprocessing_results(X, X_transformed, pipeline)
```

```python
#ФИНАЛЬНЫЙ ОТЧЕТ
def generate_final_report(original_df, processed_data, pipeline):
    """Генерация финального отчета о предобработке"""

    print("=" * 60)
    print("ФИНАЛЬНЫЙ ОТЧЕТ О ПРЕДОБРАБОТКЕ ДАННЫХ")
    print("=" * 60)

    print(f"\n1. ОБЩАЯ ИНФОРМАЦИЯ:")
    print(f"   - Исходная размерность: {original_df.shape}")
    print(f"   - Размерность после обработки: {processed_data.shape}")
    print(f"   - Тип данных после обработки: {type(processed_data)}")

    print(f"\n2. ОБРАБОТКА ПРОПУЩЕННЫХ ЗНАЧЕНИЙ:")
    missing_before = original_df.isnull().sum().sum()
    print(f"   - Пропусков до обработки: {missing_before}")
    print(f"   - Пропусков после обработки: 0")

    print(f"\n3. ПРЕОБРАЗОВАНИЕ ТИПОВ ДАННЫХ:")
    numeric_before = original_df.select_dtypes(include=[np.number]).shape[1]
    categorical_before = original_df.select_dtypes(include=['object']).shape[1]
    print(f"   - Числовых признаков до: {numeric_before}")
    print(f"   - Категориальных признаков до: {categorical_before}")
    print(f"   - Всего признаков после: {processed_data.shape[1]} (все числовые)")

    print(f"\n4. НОРМАЛИЗАЦИЯ/СТАНДАРТИЗАЦИЯ:")
    if processed_data.shape[1] > 0:
        means = np.mean(processed_data, axis=0)
        stds = np.std(processed_data, axis=0)
        print(f"   - Средние значения признаков: ~{np.mean(means):.4f} ± {np.std(means):.4f}")
        print(f"   - Стандартные отклонения: ~{np.mean(stds):.4f} ± {np.std(stds):.4f}")

    print(f"\n5. ИСКУССТВЕННЫЕ ОШИБКИ:")
    # Используем обновленную функцию с аргументом
    errors = generate_artificial_errors(original_df)
    if errors:
        for error in errors:
            print(f"   ⚠️ {error}")
    else:
        print("   - Все этапы предобработки были выполнены успешно")

    print(f"\n6. ПРИМЕР ПРЕОБРАЗОВАННЫХ ДАННЫХ:")
    print(f"   Первые 3 строки преобразованных данных:")
    sample_data = pd.DataFrame(processed_data[:3])
    if hasattr(pipeline, 'transformed_feature_names') and pipeline.transformed_feature_names is not None:
        # Убедимся, что количество имен признаков совпадает с размерностью данных
        if len(pipeline.transformed_feature_names) == processed_data.shape[1]:
            sample_data.columns = pipeline.transformed_feature_names
        else:
            sample_data.columns = [f'Feature_{i}' for i in range(processed_data.shape[1])]
    else:
        sample_data.columns = [f'Feature_{i}' for i in range(processed_data.shape[1])]
    print(sample_data.round(4))

# Генерируем финальный отчет
generate_final_report(df, X_transformed, pipeline)
```
